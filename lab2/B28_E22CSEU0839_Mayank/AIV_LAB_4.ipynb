```python
Lena=cv2.imread('/content/image.png')
```

```python
Gray_lena=cv2.cvtColor(Lena,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(Gray_lena,cmap='gray')
plt.title('Gray_Lena')
plt.axis('off')
plt.show()
```

```python
hist_gray=cv2.calcHist([Gray_lena],[0],None,[256],[0,256])
plt.plot(hist_gray)
plt.title('Gray_Lena_Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
color_lena=cv2.cvtColor(Lena,cv2.COLOR_BGR2RGB)
plt.imshow(color_lena)
plt.title('Color_Lena')
plt.axis('off')
plt.show()
```

```python
colors=('b','g','r')
for i,col in enumerate(colors):
  hist=cv2.calcHist([color_lena],[i],None,[256],[0,256])
  plt.plot(hist,color=col)
plt.title('Color_Lena_Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.imshow(Gray_lena,cmap='gray')
plt.title('Gray_Lena')
plt.axis('off')
plt.show()
```

```python
f_transform=np.fft.fft2(Gray_lena)
fshift=np.fft.fftshift(f_transform)
magnitude_spectrum=20*np.log(np.abs(fshift)+1)
plt.figure(figsize=(6,6))
plt.imshow(magnitude_spectrum,cmap='gray')
plt.title('Magnitude Spectrum')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
from google.colab.patches import cv2_imshow
```

```python
f_ishift=np.fft.ifftshift(fshift)
i_transform=np.fft.ifft2(f_ishift)
i_transform=np.abs(i_transform)
cv2_imshow(i_transform.astype('uint8'))
```

```python
angle=45
rotated=cv2.warpAffine(Gray_lena,cv2.getRotationMatrix2D((Gray_lena.shape[1]//2,Gray_lena.shape[0]//2),angle,1),(Gray_lena.shape[1],Gray_lena.shape[0]))
cv2_imshow(rotated)
```

```python
forier_rotated=np.fft.fft2(rotated)
fshift_rotated=np.fft.fftshift(forier_rotated)
magnitude_spectrum_rotated=20*np.log(np.abs(fshift_rotated)+1)
plt.figure(figsize=(6,6))
plt.imshow(magnitude_spectrum_rotated,cmap='gray')
plt.title('Magnitude Spectrum')
plt.xticks([]),plt.yticks([])
plt.show()
```
