```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Task-1: 2D Transformations on a Triangle
vertices = np.array([[0, 0], [4, 0], [2, 3]])  # Triangle vertices
vertices = np.append(vertices, [vertices[0]], axis=0)  # Close the triangle
plt.plot(vertices[:, 0], vertices[:, 1], marker='o')
plt.title("Original Triangle")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Translation
translation_matrix = np.array([[1, 0, 3], [0, 1, 2], [0, 0, 1]])
vertices_homogeneous = np.hstack([vertices, np.ones((vertices.shape[0], 1))])
translated_vertices = (translation_matrix @ vertices_homogeneous.T).T[:, :2]
plt.plot(translated_vertices[:, 0], translated_vertices[:, 1], marker='o')
plt.title("Translated Triangle")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Scaling
scaling_matrix = np.array([[2, 0, 0], [0, 1.5, 0], [0, 0, 1]])
scaled_vertices = (scaling_matrix @ vertices_homogeneous.T).T[:, :2]
plt.plot(scaled_vertices[:, 0], scaled_vertices[:, 1], marker='o')
plt.title("Scaled Triangle")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Rotation
angle_rad = np.radians(45)
rotation_matrix = np.array([
    [np.cos(angle_rad), -np.sin(angle_rad), 0],
    [np.sin(angle_rad), np.cos(angle_rad), 0],
    [0, 0, 1]
])
rotated_vertices = (rotation_matrix @ vertices_homogeneous.T).T[:, :2]
plt.plot(rotated_vertices[:, 0], rotated_vertices[:, 1], marker='o')
plt.title("Rotated Triangle")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Reflection
reflection_matrix = np.array([[1, 0, 0], [0, -1, 0], [0, 0, 1]])
reflected_vertices = (reflection_matrix @ vertices_homogeneous.T).T[:, :2]
plt.plot(reflected_vertices[:, 0], reflected_vertices[:, 1], marker='o')
plt.title("Reflected Triangle (X-Axis)")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Shearing
shearing_matrix = np.array([[1, 1.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_vertices = (shearing_matrix @ vertices_homogeneous.T).T[:, :2]
plt.plot(sheared_vertices[:, 0], sheared_vertices[:, 1], marker='o')
plt.title("Sheared Triangle")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Loading the image
img = cv2.imread("/content/Picture1.jpg")  # Replace with your image path
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
# Image Translation
rows, cols = img.shape[:2]
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
translated_img = cv2.warpAffine(img, translation_matrix, (cols, rows))
plt.imshow(translated_img)
plt.title("Translated Image")
plt.show()
```

```python
# Image Scaling
scaled_img = cv2.resize(img, None, fx=0.2, fy=0.2)
plt.imshow(scaled_img)
plt.title("Scaled Image")
plt.show()
```

```python
# Image Rotation
center = (cols / 2, rows / 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (cols, rows))
plt.imshow(rotated_img)
plt.title("Rotated Image")
plt.show()
```

```python
# Image Reflection
reflected_img = cv2.flip(img, 1)  # Horizontal flip
plt.imshow(reflected_img)
plt.title("Reflected Image")
plt.show()
```

```python
# Image Cropping
cropped_img = img[50:200, 50:200]
plt.imshow(cropped_img)
plt.title("Cropped Image")
plt.show()
```

```python
# Image Shearing
shear_matrix = np.float32([[1, 0.5, 0], [0, 1, 0]])
sheared_img = cv2.warpAffine(img, shear_matrix, (cols, rows))
plt.imshow(sheared_img)
plt.title("Sheared Image")
plt.show()
```
