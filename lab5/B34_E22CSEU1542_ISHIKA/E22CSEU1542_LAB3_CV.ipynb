```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow  # Import cv2_imshow

# Task-1: Image Manipulation
img = cv2.imread('/content/Lena_Image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display Image
plt.imshow(img_rgb)
plt.title('Original Image')
plt.show()

# Extract Image Size
img_height, img_width, img_channels = img.shape
print(f'Image Size: {img_width}x{img_height}, Channels: {img_channels}')

# Calculate Total Pixels
total_pixels = img_height * img_width
print(f'Total Pixels: {total_pixels}')

# Convert to Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)

# Convert to Binary using Threshold
_, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(binary_img)

# Count Black Pixels
black_pixel_count = np.sum(binary_img == 0)
print(f'Black Pixels Count: {black_pixel_count}')

# Task-2: Edge Detection
sobel_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = cv2.magnitude(sobel_x, sobel_y)
cv2_imshow(sobel_edges)

prewitt_x = cv2.filter2D(gray_img, -1, np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]]))
prewitt_y = cv2.filter2D(gray_img, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
prewitt_x = np.float32(prewitt_x)
prewitt_y = np.float32(prewitt_y)

prewitt_edges = cv2.magnitude(prewitt_x, prewitt_y)
cv2_imshow(prewitt_edges)

roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])
roberts_x_edge = cv2.filter2D(gray_img, -1, roberts_x)
roberts_y_edge = cv2.filter2D(gray_img, -1, roberts_y)
roberts_x_edge = np.float32(roberts_x_edge)
roberts_y_edge = np.float32(roberts_y_edge)
roberts_edges = cv2.magnitude(roberts_x_edge, roberts_y_edge)
cv2_imshow(roberts_edges)

canny_img = cv2.Canny(gray_img, 100, 200)
cv2_imshow(canny_img)
```
